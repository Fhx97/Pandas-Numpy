# 1.删除重复元素
  
  使用duplicated()函数检测重复的行，返回元素为布尔类型的Series对象，每个元素对应一行，如果该行不是第一次出现，则元素为True。
  
  - 使用drop_duplicats()函数删除重复的行
  - 使用duplicated()函数查看重复的行
  
  如果使用pd.concat([df1,df2],axis=1)生成新的DataFrame，新的df中columns相同，使用duplicate()与drop_duolicates()都会出现问题。

```python
# 原始数据处理一般都需要进行如下处理：空值、重复值、异常值。
df_1 = DataFrame(data={
    "name":["lucy","tom","jack","tony","mery","rose","black"],
    "python":np.random.randint(0,100,size=(7)),
    "java":np.random.randint(0,100,size=(7)),
    "php":np.random.randint(0,100,size=(7))
})
df_1
```

```python
df_1.loc[7]=df_1.loc[0].copy()
df_1
```

```python
df_1.duplicated()
```

```python
df_1.drop_duplicates()
```

# 2.映射
  
  映射的含义：创建一个映射关系列表，把values元素和一个特定的标签或者字符串绑定。包含三种操作：
  
  - replace()函数：替换元素（DataFrame\Series的函数）
  - map()函数：最重要，新建一列（Series的函数）
  - rename()函数：替换索引（DataFrame的函数）
  

```python
# 可接收列表，进行多个元素替换
df_1.replace('tom','TOM')
```

```python
# 正则表达式替换
df_1.replace(to_replace=r't.*',regex=True,value="ContainT")
```

## map()函数：新建一列。
   
   map()是Series的函数，所以通常被用来对某一列来进行整体的映射处理。  
   - map()可以使用字典映射新一列数据
   - map()中可以使用lambd表达式
   - map()中可以使用方法，可以使用自定义的方法。
   
【注意】：  
   - map()中不能使用sun之类的函数，for循环
   - map(字典)字典的键要足以匹配所有的数据，否则出现NaN。
   
   

```python
map_dict = {
    "lucy":"北京",
    "tom":"上海",
    "jack":"江西",
    "tony":"福建",
    "mery":"四川",
    "rose":"广西",
    "black":"云南"}
```

```python
df_1["address"]=df_1.name.map(map_dict)
```

```python
df_1
```

```python
def score_fhx( score):
    if  score >= 90:
        return "A"
    elif  score >= 80:
        return "B"
    elif  score >= 70:
        return "C"
    elif score >= 60:
        return "D"
    else:
        return "E"
```

```python
df_1.java.map(score_fhx)
```

```python
#lambda表达式
df_1.name.map(lambda x: x + "一班")
```

## rename()函数：替换索引

  仍然是新建一个字典，使用rename()函数替换行索引。  
  
  - mapper 替换所有索引
  - index 替换行索引
  - columns 替换列索引
  - level 指定多维索引的维度


```python
df_1.rename(index={"lucy":"张三","tom":"李四"})
```

  - 异常值通用界定方法：如果数据呈标准正态分布，|data| > 3 * data.std() ====》异常值。  
  - 离群点检测：IQR=Q3-Q1  下边界=Q1-1.5*IQR   上边界=Q3+1.5*IQR
  

```python
data = np.random.randn(1000)
data
```

```python
#将数据转换为绝对值形式
data[np.abs(data) > 3*data.std()]
```

```python
df_4 = DataFrame(data=np.random.randn(1000,3),columns=list("ABC"))
df_4
```

```python
# 任意一行至少存在一个数的绝对值大于该数所处列的三倍标准差，即认定为满足异常值条件。
con_1= (np.abs(df_4)>3*df_4.std()).any(axis=1)
```

```python
df_4.loc[con_1]
```

```python
#删除异常值
df_4.drop(df_4.loc[con_1].index)
```
